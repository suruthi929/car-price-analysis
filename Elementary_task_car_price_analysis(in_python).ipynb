```python
#  Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Make plots appear inline in notebook
%matplotlib inline
```

```python
# Step 3: Load the Data
df = pd.read_csv("CarPrice_Assignment.csv")
df.head()  # Display first 5 rows
```

```python
# Step 4: Inspect the Data
df.info()
df.describe()
```

```python
# Step 5: Check Null Percentage
null_percent = (df.isnull().sum() / len(df)) * 100
print("Null Percentage in Each Column:\n", null_percent)
```

```python
# Step 6: Remove Columns with High Nulls (if any)
df = df.loc[:, null_percent < 40]  # Keep only columns with <40% nulls
```

```python
# Step 7: Remove Rows with Too Many Nulls (if any)
df = df[df.isnull().sum(axis=1) < 5]  # Drop rows with >=5 nulls
```

```python
# Step 8: Drop Unwanted Columns
df.drop(columns=['car_ID'], inplace=True)

```

```python
# Step 9: Sort by Price (optional)
df.sort_values(by='price', ascending=False).head()
```

```python
# Step 10: Pairplot for Some Key Relationships
sns.pairplot(df[['price', 'horsepower', 'enginesize', 'curbweight']])
plt.show()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
